#### For the Age of the passangers ... how would you guess at the missing values using the other data present in the CSV.

We could look at subsets of the passengers and hope to get meaningful information this way. For example, we can use masking to obtain a subset of our entries where the name (in the 'name' column) includes the word 'Miss'. This subset, based on the corresponding sibsp and parch entries, seem to refer to females that were either children or single, unmarried women:

```python
titanic_db_with_just_miss = titanic[titanic['name'].str.contains('Miss')]
titanic_db_with_just_miss.head(10)
```

We can also use a pivot table to group the passengers, and then find the mean ages of those groups.
For example, we can use a pivot table to group the passengers by passenger class (the 'pclass column'), find the mean age of each passenger group, and then use that mean to fill in the missing values in the age column, based on that passenger's class.

```python
means_of_groups_by_pclass = pd.pivot_table(titanic, index='pclass', values=['age'],
                                           aggfunc=[np.mean])
means_of_groups_by_pclass
```

Or we can use a pivot table and group the passengers by gender (the 'sex' column) to obtain the mean of each gender group:

```python
means_of_groups_by_gender = pd.pivot_table(titanic, index='sex', values=['age'],
                                           aggfunc=[np.mean])
means_of_groups_by_gender
```

I am interested in studying the survival rate differences between the genders, so I have decided to use my 2nd pivot table (based on grouping by gender) to fill in the missing values in the 'age' column:

```python
# First, we use our mean of 30.59 (for males) to fill in missing age values for the men:
titanic.loc[(titanic['sex'] == 'male') & (titanic['age'].isnull()), 'age'] = 30.59

# Next, we use our mean of 28.69 (for females) to fill in missing age values for the women:
titanic.loc[(titanic['sex'] == 'female') & (titanic['age'].isnull()), 'age'] = 28.69
titanic.head(60)
```

## Part 4: Group Statistics

#### Are there any groups that were especially adversely affected in the Titanic wreck? (justify your response numerically)

```python
# I will start by making a column called 'deceased' that assigns a 1 if the person died,
# or a 0 if the person survived. By doing this, I can obtain a count of how many people died
# and then use this information in my pivot tables later
titanic['deceased'] = titanic['survived'].apply(lambda x: 1 if x == 0 else 0)

# I will also make a total column, which is my 'survived' + 'deceased' column, and will aid
# me in having a total value for each group later for my pivot tables
titanic['total'] = titanic['survived'] + titanic['deceased']
titanic['survivalRate'] = titanic['survived']
```

```python
groups_by_class_and_gender = pd.pivot_table(titanic, index=['pclass', 'sex', 'embarked'], 
                                  values=['survived','deceased', 'survivalRate','total'],
                                           
                                            aggfunc={'deceased': np.sum,
                                                     'survived': np.sum,
                                                     'total': 'count',
                                                     'survivalRate': np.mean}
                                           )

groups_by_class_and_gender
```

```python
count_of_people_who_survived_by_gender = pd.pivot_table(titanic, index='sex',
                                                        values=['survived','deceased',
                                                                'total'],
                                                        aggfunc=[np.sum], margins=True,
                                                        margins_name='All')
count_of_people_who_survived_by_gender
```

From viewing our pivot tables, it becomes clear that men, in general, were adversely affected. If we look at the total counts, only 161 out of a total of 843 men survived. If we begin to
factor in passenger class, men in the 2nd passenger class only had either a 29% or 13% survival rate (depending on their embarked point). Men in the 3rd passenger class fared even worse: survival rates were either 21%, 12%, or 14%, depending on their embarked point.

#### Are there any groups that outperformed the survival of the latter group? (justify your response numerically)


```python
# We will use a Chi-Squared test to compare females to males. We will use our previous
# pivot table

# expected value for females, deceased
ev_females_deceased = (809./1309) * 466

# expected value for males, deceased
ev_males_deceased = (809./1309) * 843

# expected value for females, survived
ev_females_survived = (500./1309) * 466

# expected value for males, survived
ev_males_survived = (500./1309) * 843

observed = [127, 682, 339, 161]
expected = [ev_females_deceased, ev_males_deceased, ev_females_survived, ev_males_survived]
```

```python
from scipy.stats import chisquare
chisquare(observed, f_exp=expected)
```

From our Chi-Squared Test, we obtain a pvalue that is nearly 0, so it is clear that there was
indeed a significant difference between male and female survival rates on the Titanic, and we
may reject the null hypothesis. We can also see quite a difference between the genders visually, as shown below. With 0 (in blue) representing "died", and 1 (in green) representing "survived", it is evident that women fared far better than men on the Titanic.

```python
titanic_groups = titanic.groupby('sex')['survived'].value_counts().sort_index()
titanic_groups.unstack().plot(kind='bar', subplots=True, layout=(1,2))
plt.show()
```

## Part 5:  Comparative Statistics:  Lusitania

#### Are there any groups that were especially adversely affected in the Lusitania wreck? (justify your response numerically)

```python
lusitania = pd.read_csv('assets/lusitania.csv')
lusitania.head(50)
```

```python
# check the data dtypes for each column in Lusitania
lusitania.dtypes
```

```python
# We'll work on our age column using a temporary 'Age_temp' column
lusitania['Age_temp'] = lusitania['Age']
```

```python
lusitania['Age_temp'].unique()
```

```python
# Clean up the age values
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('?', '')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('Infant', '1')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('(\d\d)-months', '1')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('(\d.\d+)-months', '1')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('_', '5')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('3-6', '4.75')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('51 or 47', '49')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('45 or 51', '48')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('\(\)', '')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('\d\d \(\d\d\)', '')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('\d? \(\d\)', '')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace('0\d', '')
lusitania['Age_temp'] = lusitania['Age_temp'].str.replace(' ', '')

lusitania['Age_temp'].unique()
```

```python
# Let us check the number of null values in the age column
lusitania['Age_temp'].isnull().sum()
```

```python
# reassign our temporary 'Age_temp' column to our original 'Age' column, and then drop
# the 'Age_temp' column
lusitania['Age'] = lusitania['Age_temp']
lusitania.drop('Age_temp', inplace=True, axis=1)
lusitania.head()
```

Because of the large number of missing age values, and also because I am interested in comparing gender differences between both the Titanic and the Lusitania, I have decided to focus on male vs female survival rates on the Lusitania. I will start by inspecting the 'Fate' column.

```python
lusitania['Fate'].unique()
```

```python
# Here, I am making 'Saved', 'Lost', 'Saved (died from trauma)', 'Not on board',
# 'Total', and 'SurvivalRate' columns for my pivot table later
lusitania['Saved'] = lusitania['Fate'].apply(lambda x: 1 if x == "Saved" else 0)
lusitania['Lost'] = lusitania['Fate'].apply(lambda x: 1 if x == "Lost" else 0)
lusitania['Saved (died from trauma)'] = lusitania['Fate'].apply(lambda x: 1 if x == "Saved (died from trauma)" else 0)
lusitania['Not on board'] = lusitania['Fate'].apply(lambda x: 1 if x == "Not on board" else 0)
lusitania['Total'] = lusitania['Saved'] + lusitania['Lost']
lusitania['SurvivalRate'] = lusitania['Saved']
lusitania.head()
```

```python
lusitania_pivot_table = pd.pivot_table(lusitania, index=['Sex', 'Passenger/Crew', 'Department/Class'],
                                                        values=['Saved','Lost',
                                                                'Saved (died from trauma)',
                                                                'Not on board', 'Total',
                                                                'SurvivalRate'],
                                                        aggfunc={'Saved': np.sum,
                                                                 'Lost': np.sum,
                                                                 'Saved (died from trauma)': np.sum,
                                                                 'Not on board' : np.sum,
                                                                 'Total' : 'count',
                                                                 'SurvivalRate' : np.mean}, margins=True,
                                                        margins_name='All')
lusitania_pivot_table
```

```python
# Here, I am making a crosstable to use for my Chi-Squared Test
crosstable_lusitania_gender = pd.crosstab(lusitania['Sex'], lusitania['Saved'])
crosstable_lusitania_gender
```

```python
from scipy.stats import chi2_contingency
chisquared_value, pvalue_gender_survival, df, expected = chi2_contingency(crosstable_lusitania_gender)
print "Chi-Squared Value is:", chisquared_value
print "Pvalue: ", pvalue_gender_survival
```

From our Chi-Squared Test, we obtain a Pvalue of .29, so it is clear that there was NOT a
significant difference between male and female survival rates on the Lusitania, and we fail
to reject the null hypothesis.

#### Are there any groups that outperformed the survival of the latter group? (justify your response numerically)



```python
# Set up another crosstable for Chi-Squared test between Passengers and Crew members
crosstable_lusitania_passengerCrew = pd.crosstab(lusitania['Passenger/Crew'], lusitania['Saved'])
crosstable_lusitania_passengerCrew
```

```python
chisquared_value, pvalue_pass_survival, df, expected = chi2_contingency(crosstable_lusitania_passengerCrew)
print "Chi-Squared Value is:", chisquared_value
print "Pvalue: ", pvalue_pass_survival
```

Our pvalue of .059 is very close to our threshold of .05 that is required to reject the null hypothesis. So, though we fail to reject the null hypothesis, there does appear to be at least some difference in survival rates between Crew and Passengers.

```python
lusitania_groups = lusitania.groupby('Passenger/Crew')['Saved'].value_counts().sort_index()
lusitania_groups.unstack().plot(kind='bar', subplots=True, layout=(1,3))
plt.show()
```

#### What does the group-wise survival rate imply about circumstances during these two accidents?

In the case of the Titanic, it is incredibly clear that if you were a female on that ship, your chances of survival was significantly better than if you were a male. As for the Lusitania, though there were no gender differences, there was some difference, one that was close to being statistically signifcant, between passenger and crew survival rates.
