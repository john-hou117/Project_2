Who is using all the water?

```python
import pandas as pd
```

```python
econ_data = 'data/la_econ_data.csv'
econ_data_df = pd.read_csv(econ_data)
```

```python
econ_data_df.head()
```

```python
# let us drop Unnamed column
econ_data_df.drop('Unnamed: 0', axis=1, inplace=True)
econ_data_df.head()
```

```python
# let us replace commas in the pop column. Remember, pop is in strings, so we need to use
# the .str
# Also, and this is very CRITICAL: copy it BACK to your original. For now, let us assign our
# in progress cleaning of pop column to pop_tmp
econ_data_df['pop_tmp'] = econ_data_df['pop'].str.replace(',','')
```

```python
# we will run the .astype(float) to see what kind of error we get
econ_data_df['pop_tmp'].astype(float)
```

```python
# a function to find strange values in pop_tmp column
def try_to_cast_as_float(value):
    try:
        value = float(value)
    except ValueError:
        print (value)
```

```python
# this is pure diagnosis. We are just diagnosing the problem. This is just CATCHING the
# problem. We didn't assign the column back
econ_data_df['pop_tmp'].apply(try_to_cast_as_float)
```

```python
econ_data_df['pop_tmp'].str.replace( 'foo','').astype(float)
```

```python
# CRITICAL: let us assign/put it back into the original dataframe
econ_data_df['pop'] = econ_data_df['pop_tmp'].str.replace( 'foo','').astype(float)
```

```python
# let us get rid of pop_tmp now, since we are done with it, by dropping it
econ_data_df.drop('pop_tmp', axis=1, inplace=True)
```

```python
econ_data_df.head(3)
```

```python
# now, let us diagnose another column, the 'zip' column
```

```python
econ_data_df['zip'].apply(try_to_cast_as_float)
```

```python
# again, let us create a temp column to work with the zip
econ_data_df['zip_tmp'] = econ_data_df['zip']
```

```python
# we are using regex with captured groups, here denoted by the () around the \d\d\d\d \? \?
# in the 1st part of my replace, and then referencing that captured group by using the \\1
# in the second part of my replace
econ_data_df['zip_tmp'] = econ_data_df['zip'].str.replace('(\d\d\d\d \? \?)', '9\\1')
```

```python
# NOW, assign 'zip_tmp' back to original data, which is 'zip'
econ_data_df['zip'] = econ_data_df['zip_tmp']

# and drop the zip_tmp column now
econ_data_df.drop('zip_tmp', inplace=True, axis=1)
```

```python
econ_data_df.head()
```

```python
# now, x_bar column
# first, make temp column, x_bar_tmp. Then go after ZERO
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar'].str.replace('ZERO', '0')
```

```python
# now, the ocho. Notice we are working on x_bar_tmp
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('ocho', '8')

# now, the zwei
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('zwei', '2')

# now, the quatre
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('quatre', '4')

# now, the SAN
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('SAN', '3')

# now, the ZEVEN (important that we do this before the V, which is next)
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('ZEVEN', '7')

# now, the V
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('V', '5')

# now, the I
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('I', '1')

# now, the 3*2. We have to escape
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('\(3\*2\)', '6')

# now, the 3**2. We have to escape certain characters using \
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace('\(3\*\*2\)', '9')

# now, the space
econ_data_df['x_bar_tmp'] = econ_data_df['x_bar_tmp'].str.replace(' ', '')

# find anything else wrong. We ran this below and found two rows with .
# econ_data_df['x_bar_tmp'].apply(try_to_cast_as_float)

# so now, do a mask looking for the . character. Remember, mask is good for looking at data
econ_data_df[econ_data_df['x_bar_tmp'] == '.']
```

```python
import numpy as np

# we are going to get rid of index position 396, so we will set it to -99999999, so we can
# later assign it to np.nan
econ_data_df['x_bar_tmp'].ix[61] = '0'
econ_data_df['x_bar_tmp'].ix[396] = '-99999999'
```

```python
econ_data_df['x_bar_tmp'].iloc[396] = np.nan
```

```python
# here, we are dropping na AND also we are assigning it back to the original
econ_data_df = econ_data_df.dropna()

# assign x_bar_tmp back to x_bar
econ_data_df['x_bar'] = econ_data_df['x_bar_tmp']

# and drop the x_bar_tmp column
econ_data_df.drop('x_bar_tmp', axis=1, inplace=True)
```

```python
econ_data_df.head()
```

```python
# let us rename the columns
econ_data_df.columns = ['Zip Code', 'Mean', 'Population', 'Median']
```

```python
econ_data_df.head()
```

```python
# now, the Median column
# we made a Median_tmp column, and now we are replacing 'and' with '', and also ',' with ''
econ_data_df['Median_tmp'] = econ_data_df['Median'].str.replace('and','')
econ_data_df['Median_tmp'] = econ_data_df['Median_tmp'].str.replace(',','')

econ_data_df.head()
```

```python
# extract numbers. We are splitting on the word 'thous'
econ_data_df['Median_thousands'] = econ_data_df['Median_tmp'].str.extract('(.+) thous',
                                                                        expand=False)

econ_data_df['Median_ones'] = econ_data_df['Median_tmp'].str.extract('thous (.+)',
                                                                        expand=False)
# drop some of our old work
econ_data_df.drop(['Median_mega', 'Median_kilo'], axis=1, inplace=True)
```

```python
econ_data_df.head(10)
```

```python
# and so on, and so on, and so on, with cleaning
```

```python
# let us make a new DataFrame so we can practice joining
la_econ_data = pd.DataFrame([econ_data_df['Zip Code'], 
              econ_data_df['Mean'].astype(float), 
              econ_data_df['Population'].astype(int)]).T
la_econ_data.head(30)
```

```python
la_water_data = pd.read_csv('data/Water_Use_Average_By_Zipcode.csv')
la_water_data.head(5)
```

```python
# clean the location column. Notice the \\n, not just \n, because there is actually a \n
la_water_data['Zip Code'] = la_water_data['Location 1'].str.extract('(\d\d\d\d\d)\\n',
                                                              expand=False)
```

```python
# Drop the Location 1 column
la_water_data.drop('Location 1', axis=1, inplace=True)
la_water_data.head()
```

```python
# merge the data
la_water_data = la_water_data.merge(la_econ_data, how='inner')
```

```python
la_water_data['Aggregate Water Usage'] = (la_water_data['FY 05/06'] + la_water_data['FY 06/07']
                                        + la_water_data['FY 07/08']
                                        + la_water_data['FY 08/09']
                                        + la_water_data['FY 09/10']
                                        + la_water_data['FY 10/11']
                                        + la_water_data['FY 11/12']
                                        + la_water_data['FY 12/13'])
la_water_data.head()
```
